### Test Inference

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
TOKEN = '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'

# Load the model and tokenizer
# path = 'fajrikoto/llama3p1_8b_extend25p_3-1-3_CL_RT_en'
# path = 'fajrikoto/llama3_3p1_8b_extend25p_3-1-3_CL_RT_en_IFT_2.7M_orca_Nanda_en'
path = 'fajrikoto/llama3_3p1_8b_extend25p_3-1-3_CL_RT_en_IFT_final'


tokenizer = AutoTokenizer.from_pretrained(path, token=TOKEN)
model = AutoModelForCausalLM.from_pretrained(path, token=TOKEN, 
                                             attn_implementation="flash_attention_2", 
                                             torch_dtype=torch.bfloat16, device_map="auto")
device = model.device # Get the device the model is loaded on
# Define conversation input

tokenizer.chat_template="{% set loop_messages = messages %}{% for message in loop_messages %}{% set content = '<|start_header_id|>' + message['role']+'<|end_header_id|>\n\n'+ message['content'] | trim + '<|eot_id|>' %}{% if loop.index0 == 0 %}{% set content = bos_token + content %} {% endif %}{{ content }}{% endfor %}{% if add_generation_prompt %}{{ '<|start_header_id|>assistant<|end_header_id|>\n\n' }}{% endif %}"

conversation = [
    {"role": "user", "content": "Give a recipe for making a salad."}
]

input_ids = tokenizer.apply_chat_template(
    conversation=conversation,
    tokenize=True,
    add_generation_prompt=True,
    return_tensors="pt").to(device)

# Generate a response
gen_tokens = model.generate(
    input_ids,
    max_new_tokens=100,
    stop_strings=["<|eot_id|>"],
    tokenizer=tokenizer
    )

# Decode and print the generated text along with generation prompt
gen_text = tokenizer.decode(gen_tokens[0][len(input_ids[0]): -1])
print(gen_text)
```
