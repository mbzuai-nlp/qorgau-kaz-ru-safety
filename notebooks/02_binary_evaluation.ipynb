```python
safe_count = safe_labels.count(False)
unsafe_count = safe_labels.count(True)
total = len(safe_labels)

print(f"Safe: {safe_count}")
print(f"Unsafe: {unsafe_count}")
print(f"Percentage safe: {safe_count / total * 100:.2f}%")
```
