```python
def construct_rubric(rubric_key: str, lang: str) -> str:
    """
    Construct the rubric text based on the rubric_fine_grained dictionary.

    Args:
        rubric_key (str): Either "safe" or "unsafe".

    Returns:
        str: A formatted string with rubric items.
    """
    if lang == 'ru':
        rubric_fine_grained = rubric_ru_fine_grained
    else:
        rubric_fine_grained = rubric_kz_fine_grained
                              
    if rubric_key not in rubric_fine_grained:
        raise ValueError(f"Invalid rubric key: {rubric_key}. Must be 'safe' or 'unsafe'.")
    
    rubric_dict = rubric_fine_grained[rubric_key]
    
    rubric_text = "\n".join([f"{idx}. {desc}" for idx, desc in rubric_dict.items()])
    return rubric_text
```

```python
import pandas as pd

def extract_responses(batch_response_jsonfile):
    df = pd.read_json(batch_response_jsonfile, lines=True)
    # Extract and sort by custom_id
    ids = [int(cid.split("-")[-1]) for cid in df['custom_id']]
    df['ids'] = ids
    df_sorted = df.sort_values(by='ids', ascending=True).reset_index(drop=True)
    
    # Collect responses
    responses = []
    for _, row in df_sorted.iterrows():
        result = ''.join(choice['message']['content'] for choice in row['response']['body']['choices'])
        responses.append(result)
    
    assert len(df) == len(responses)
    return responses

def parse_and_add_labels(df, file_path, language="KZ", ):
    """
    Process responses, parse labels, and add to the DataFrame.

    Args:
        df (pd.DataFrame): The main DataFrame (e.g., df_kz or df_ru).
        data_paths (dict): A dictionary where keys are model names and values are JSONL file paths.
        label_column_prefix (str): Prefix for label column names (e.g., 'aya_fg_label').
        language (str): Either "KZ" or "RU" to differentiate.
    """
    responses = extract_responses(file_path)
    labels = parse_labels(responses)
    
        # Add the labels as a new column
    column_name = "gpt_finegrained"
    df[column_name] = [" ".join(map(str, label_list)) for label_list in labels]
    # Increment labels by 1
    df[column_name] = df[column_name].astype(int) + 1

def parse_labels(reviews):
    """
    Parse labels from reviews where answers are in the format:
    <answer>index</answer>.
    
    Args:
        reviews (list): List of response strings containing <answer> tags.
        
    Returns:
        list: List of parsed labels, where each label is a list of integers.
    """
    labels = []

    for r in reviews:
        if "<answer>" in r and "</answer>" in r:
            # Extract the content between <answer> and </answer>
            temp = r.split("<answer>")[1].split("</answer>")[0].strip()
            
            try:
                label_list = [int(index.strip()) for index in temp.split()]
                labels.append(label_list)
            except ValueError:
                labels.append([-1])
        else:
            labels.append([-1])

    return labels

```
